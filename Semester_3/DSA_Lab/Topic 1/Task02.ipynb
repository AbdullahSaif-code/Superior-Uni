````python
import time
import matplotlib.pyplot as plt

# Recursive approach
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Iterative approach
def fibonacci_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Optimized recursive approach using memoization
def fibonacci_memoized(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_memoized(n - 1, memo) + fibonacci_memoized(n - 2, memo)
    return memo[n]

# Measure execution time for each approach and visualize
def measure_execution_time_and_plot():
    ns = [10, 20, 30, 40]
    execution_times = {"Recursive": [], "Iterative": [], "Memoized": []}

    for n in ns:
        # Recursive approach
        start_time = time.time()
        fibonacci_recursive(n)
        end_time = time.time()
        execution_times["Recursive"].append(end_time - start_time)

        # Iterative approach
        start_time = time.time()
        fibonacci_iterative(n)
        end_time = time.time()
        execution_times["Iterative"].append(end_time - start_time)

        # Memoized approach
        start_time = time.time()
        fibonacci_memoized(n)
        end_time = time.time()
        execution_times["Memoized"].append(end_time - start_time)

    # Plot execution times
    for approach, times in execution_times.items():
        plt.plot(ns, times, label=approach)

    plt.xlabel('n (Fibonacci Number Index)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Execution Time Comparison of Fibonacci Approaches')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the execution time measurement and visualization
measure_execution_time_and_plot()
````
