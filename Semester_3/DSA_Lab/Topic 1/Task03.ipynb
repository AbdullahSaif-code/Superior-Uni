````python
import numpy as np

import matplotlib.pyplot as plt

# Define functions for different complexities
def constant(n):
    return np.ones_like(n)

def logarithmic(n):
    return np.log2(n)

def linear(n):
    return n

def linearithmic(n):
    return n * np.log2(n)

def quadratic(n):
    return n ** 2

def exponential(n):
    return 2 ** n

# Generate input sizes
n = np.arange(1, 1001)

# Compute output values for each complexity
y_constant = constant(n)
y_logarithmic = logarithmic(n)
y_linear = linear(n)
y_linearithmic = linearithmic(n)
y_quadratic = quadratic(n)
y_exponential = exponential(n[:20])  # Limit exponential to avoid overflow

# Plot the complexities
plt.figure(figsize=(12, 8))
plt.plot(n, y_constant, label="O(1)", linewidth=2)
plt.plot(n, y_logarithmic, label="O(log n)", linewidth=2)
plt.plot(n, y_linear, label="O(n)", linewidth=2)
plt.plot(n, y_linearithmic, label="O(n log n)", linewidth=2)
plt.plot(n, y_quadratic, label="O(n^2)", linewidth=2)
plt.plot(n[:20], y_exponential, label="O(2^n)", linewidth=2)  # Plot limited exponential

# Customize the plot
plt.yscale('log')  # Use logarithmic scale for better visualization
plt.title("Growth of Different Time Complexities", fontsize=16)
plt.xlabel("Input Size (n)", fontsize=14)
plt.ylabel("Operations (log scale)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()

# Show the plot
plt.show()
````
