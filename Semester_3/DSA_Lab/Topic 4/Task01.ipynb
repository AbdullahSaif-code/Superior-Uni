# Implementing a Stack Using Arrays and Linked Lists 

```python
import time

# Stack using list
class stack_list:
    def __init__(self):
        self.stack=[]
    
    def push(self,element):
        self.stack.append(element)
    
    def pop(self):
        self.stack.pop()
    
    def peek(self):
        return self.stack[-1]
    
    def is_empty(self):
        if len(self.stack) == 0:
            print("Stack is empty.")
        else:
            print("Stack is not empty.")
    
    def size(self):
        return len(self.stack)

class stack_linked_list:
    class Node:
        def __init__(self,value):
            self.value=value
            self.next=None
    
    def __init__(self):
        self.head=None
        self._size=0
    
    def push(self,element):
        new_node=self.Node(element)
        new_node.next=self.head
        self.head=new_node
        self._size+=1
    
    def pop(self):
        if self.head is None:
            print("Stack is empty.")
            return None
        popped_value=self.head.value
        self.head=self.head.next
        self._size-=1
        return popped_value
    
    def peek(self):
        if self.head is None:
            print("Stack is empty.")
            return None
        return self.head.value
    
    def is_empty(self):
        return self.head is None
    
    def size(self):
        return self._size

def comparison():
    print("For Python List")
    start = time.time()
    obj1=stack_list()
    obj1.push(1)
    obj1.push(2)
    obj1.push(3)
    obj1.pop()
    obj1.peek()
    obj1.is_empty()
    obj1.size()
    end = time.time()
    print(f"Time Taken: {end-start: .10f} seconds.")

    print("For Linked List")
    start = time.time()
    obj2=stack_linked_list()
    obj2.push(1)
    obj2.push(2)
    obj2.push(3)
    obj2.pop()
    obj2.peek()
    obj2.is_empty()
    obj2.size()
    end = time.time()
    print(f"Time Taken: {end-start: .10f} seconds.")
comparison()
```
