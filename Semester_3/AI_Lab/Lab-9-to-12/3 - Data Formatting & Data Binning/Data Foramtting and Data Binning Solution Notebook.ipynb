# Data Binning
<b>Why binning?</b>
<p>
    Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.
</p>

```python
# lets visualize the horsepower data in histogram
df['horsepower'].hist(bins=20, color='blue', edgecolor='black')
import matplotlib.pyplot as plt
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Histogram")
plt.show()
```

```python
# Create bins.
df['horsepower'] = df['horsepower'].astype(float, copy=True)
bins = np.linspace(0, 200, 11) # 0-200 with 10 intervals
```

```python
# Create group names.
group_names = ['Low', 'Medium', 'High', 'Very High', 'Extreme', 'Ultra', 'Super', 'Mega', 'Hyper', 'Extreme+']
```

```python
# perform binning.
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']].head(20)
```

```python
# See the frequency of each bin with value_counts()
df['horsepower-binned'].value_counts().sort_index()
```

```python
# Lets visualize again with bins.
df['horsepower-binned'].value_counts().sort_index().plot(kind='bar', color='blue', edgecolor='black')
plt.xlabel("Horsepower Binned")
```
