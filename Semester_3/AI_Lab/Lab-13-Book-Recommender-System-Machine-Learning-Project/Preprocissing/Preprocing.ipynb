## Popularity Based Recommender System

-> Display Top 50 Books with hiest rating avrage <br>
-> Book shold have minimum 250 votes of ratings

```python
# mearging ratings with books on basies of ISBN
ratings_on_name = ratings.merge(books,on='ISBN')
```

```python
# Finding Num of ratings acording to book name
ratings_number = ratings_on_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratings_number.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
ratings_number
```

```python
# Finding avg of ratings acording to book name
avrage_ratings = ratings_on_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avrage_ratings.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avrage_ratings
```

#### Making new df for Poperlity

```python
popular_books = ratings_number.merge(avrage_ratings,on='Book-Title')
popular_books
```

```python
popular_books =  popular_books[popular_books['num_ratings'] >= 250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popular_books
```

```python
popular_books = popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

## Collaborative Filtering Based Recommender System

-> Get User id and rating who have rated more than 200 books <br>
-> Merge the in old df

```python
x = ratings_on_name.groupby('User-ID').count()['Book-Rating'] > 200
top_users= x[x].index
top_users
```

```python
final_rating = ratings_on_name[ratings_on_name['User-ID'].isin(top_users)]
```

```python
final_rating
```

```python
y = final_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = final_rating[final_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
piviot_table = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating').fillna(0)
piviot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(piviot_table)
```

```python
similarity_scores.shape
```

#### Recomand Book Function

```python
def recommend(book_name):
    index = np.where(piviot_table.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title'] == piviot_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('The Da Vinci Code')
```
